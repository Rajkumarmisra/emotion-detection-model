```python
def rolling_rms(signal, window_len=0.5, fs=128):
  rms = []
  window_len = int(window_len * fs)
  for i in range(0, signal.shape[-1] - window_len):
    rms.append(np.sqrt(np.mean(np.square(signal[i : i + window_len]))))
  return np.array(rms)

def get_rms(X):
  return np.sqrt(np.mean(np.square(X), axis=-1))

def get_activity(X):
  return scipy.stats.tvar(X, axis=-1)

def get_mobility(X):
  X_ = np.diff(X, axis=-1)
  return np.sqrt(get_activity(X_)/get_activity(X))

def get_complexity(X):
  X_ = np.diff(X, axis=-1)
  return get_mobility(X_)/get_mobility(X)

def get_median_frequency(P, f):
  cs = np.cumsum(P, axis=-1)
  mdfs = np.zeros(cs.shape[:-1])
  if len(P.shape) == 3:
    for t in range(mdfs.shape[0]):
      for ch in range(mdfs.shape[1]):
        mdfs[t,ch] = f[np.where(cs[t,ch,:]>np.max(cs[t,ch,:])/2)[0][0]]
  else:
    for t in range(mdfs.shape[0]):
        mdfs[t] = f[np.where(cs[t,:]>np.max(cs[t,:])/2)[0][0]]

  return mdfs

def get_spectral_moments(P, f, n=0, return_all=False, concat=False):
  if return_all:
    SMs = []
    for m in range(n + 1):
      SM = (P*(f**m)).sum(axis=-1)
      SMs.append(SM)

    if concat:
      return np.concatenate(SMs, axis=-1)
    else:
      return SMs

  else:
    return (P*(f**n)).sum(axis=-1)

def get_vcf(P, f):
  SMs = get_spectral_moments(P, f, n=2, return_all=True)
  return (SMs[2] / SMs[0]) - ((SMs[1] / SMs[0]) ** 2)

def area_under_curve(t, y):
    """
    Calculate the area under the curve defined by time and amplitude values using the trapezoidal rule.

    Parameters:
    - t: array-like, time values
    - y: array-like, amplitude values

    Returns:
    - area: float, area under the curve
    """
    area = trapz(y, dx=1/128)
    return area

# Example data (replace with your own)
t = np.linspace(0, 5, 640)
```

```python
import nitime.algorithms as tsa
from scipy.signal import find_peaks
import numpy as np
from scipy.stats import kurtosis, skew
from scipy.integrate import trapz
import matplotlib.pyplot as plt

def time_domain_features(data):

  features = []
  features.append(np.mean(data, axis=-1))    #1
  features.append(np.median(data, axis=-1))  #2
  # features.append(np.std(data, axis=-1))     #3



  # Calculate first derivative
  first_derivative = np.diff(data, axis=-1)
  mean_first_derivative = np.mean(first_derivative, axis=-1)
  std_dev_first_derivative = np.std(first_derivative, axis=-1)
  features.append(mean_first_derivative)           #4
  # features.append(std_dev_first_derivative)        #5



  # Calculate second derivative
  second_derivative = np.diff(first_derivative, axis=-1)
  mean_second_derivative = np.mean(second_derivative, axis=-1)
  std_dev_second_derivative = np.std(second_derivative, axis=-1)
  features.append(mean_second_derivative)        #6
  features.append(std_dev_second_derivative)     #7

  # Mean Absolute Value (MAV)
  mav = np.mean(np.abs(data),axis=-1)
  features.append(mav)                           #8

  # Mean Absolute Value Slope (MAVS)
  mavs = np.mean(np.abs(np.diff(data, axis=-1)),axis=-1)
  features.append(mavs)                          #9

  # Zero Crossings (ZC)
  zc = np.sum(np.diff(np.sign(data), axis=-1) != 0, axis=-1)//2
  features.append(zc)                            #10


  # Variance of EMG (VAR)
  variance = np.var(data, axis=-1)
  features.append(variance)                      #11

  # Root Mean Square (RMS)
  # rms = np.sqrt(np.mean(np.square(data),axis=-1))
  # features.append(rms)                            #12

  # Mean of Peaks (MNP)
  mnpt = np.zeros(data.shape[:1])
  numPeaks = np.zeros(data.shape[:1])
  # for i, trial in enumerate(data):
  for j, signal in enumerate(data):
    peak_indices, properties = find_peaks(signal)
    mnpt[j] = np.mean(signal[peak_indices])
    numPeaks[j] = peak_indices.size
  features.append(mnpt)                           #13

  # Skewness (skew)
  skew = scipy.stats.skew(data, axis=-1)
  features.append(skew)                            #14

  # Kurtosis (kurt)
  kurt = scipy.stats.kurtosis(data, axis=-1)
  features.append(kurt)                            #15

  # Number of Peaks (NP)
  features.append(numPeaks)                        #16

  # Hjorth Mobility (HM)
  features.append(get_mobility(data))              #17

  # Hjorth Complexity (HC)
  features.append(get_complexity(data))            #18

  # Dynamic Range (DR)
  dr = np.max(data, axis=-1) - np.min(data, axis=-1)
  features.append(dr)                             #19

  # Area under the curve
  area = area_under_curve(t, data)
  features.append(area)                           #20


  #   features.append(vcf)
    # features.append(np.mean(signal))
    # features.append(np.std(signal))
    # features.append(np.sum(np.abs(signal)))  # Area
    # features.append(np.max(signal) - np.min(signal))  # Range
    # features.append(np.max(signal))  # Maximum
    # features.append(np.min(signal))  # Minimum
    # features.append(kurtosis(signal))
    # features.append(skew(signal))
    # features.append(np.sum(signal))  # Sum
    # features.append(len(signal))  # Count
    # features.append(np.var(signal))  # Variance

    # mfd = np.mean(np.abs(signal - np.mean(signal)))
    # msd = np.mean(np.square(signal - np.mean(signal)))
    # sfd = np.mean(np.power(signal - np.mean(signal), 3))
    # ssd = np.std(np.power(signal - np.mean(signal), 3))
    # dr = np.max(signal) - np.min(signal)
    # min_p = np.min(signal)
    # max_p = np.max(signal)

    # features.append(mfd)
    # features.append(msd)
    # features.append(sfd)
    # features.append(ssd)
    # features.append(dr)
    # features.append(min_p)
    # features.append(max_p)

    # # Additional non-linear features
    # features.append(np.sum(np.square(signal)))  # Sum of squares
    # features.append(np.prod(signal))  # Product
    # features.append(np.exp(np.mean(signal)))  # Exponential of mean
    # features.append(np.log1p(np.mean(signal)))  # Logarithm of mean
    # features.append(np.mean(np.abs(np.diff(signal))))  # Mean absolute difference between consecutive values
    # features.append(np.std(np.abs(np.diff(signal))))  # Standard deviation of absolute differences

  return np.array(features).T

```

```python
from scipy.stats import kurtosis, skew
def spectral_features(data):
  features = []
  # Calculate the FFT to convert to the frequency domain
  # fft_result = np.fft.fft(signal)
  # psd = np.abs(fft_result) ** 2

  # # Frequency values corresponding to the PSD data
  sample_rate = 128  # Sample rate of your data (adjust as needed)
  #   freq = np.fft.fftfreq(len(signal), 1.0 / sample_rate)

  #   psd[0] = 0
  #   psd = psd[:psd.size//2]
  #   freq = freq[:freq.size//2]
  #   # signals = row[w:e]
  #   # signals1=signals.values
  freq, psd, nu = tsa.multi_taper_psd(data, Fs=128, jackknife=False)
#   print(freq)
#   print(freq.shape)
#   print(psd)
#   print(psd.shape)
  #   # psd=P[i,:]
  #   # freq=f
  # Median Frequency (MDF)
  features.append(get_median_frequency(psd, freq))            #1

  # Mean Frequency (MNF)
  mnf = np.sum(freq * psd, axis=-1) / np.sum(psd, axis=-1)
  features.append(mnf)                                        #2

  # Peak Frequency (PKF)
  #   peaks = find_peaks(psd, height=0)  # Adjust height threshold as needed
  pkf = freq[psd.argmax(axis=-1)]
  features.append(pkf)                                        #3

  # # Mean Power (MNP)
  # mnp = psd.sum(axis=-1) / freq.size
  # features.append(mnp)                                        #4

  # # Total Power (TTP)
  # ttp = np.sum(psd, axis=-1)
  # features.append(ttp)                                        #5

  # 1st, 2nd, and 3rd Spectral Moments (SM1, SM2, SM3)
  sm1 = np.sum(freq * psd, axis=-1)
  sm2 = np.sum(np.square(freq) * psd, axis=-1)
  sm3 = np.sum(np.power(freq, 3) * psd, axis=-1)
  features.append(sm1)                                        #6
  features.append(sm2)                                        #7
  features.append(sm3)                                        #8

  # Calculate Variance of Central Frequency (VCF)
  vcf = get_vcf(psd, freq)
  features.append(vcf)                                        #9

  # Mean of Peaks frequency (MPKf)
  mnpf = np.zeros(psd.shape[:1])
  numPeaks = np.zeros(psd.shape[:1])
  # for i, trial in enumerate(psd):
  for j, signal in enumerate(psd):
    peak_indices, properties = find_peaks(signal)
    mnpf[j] = np.mean(signal[peak_indices])
    numPeaks[j] = peak_indices.size
  features.append(mnpf)                                       #10

  # Skewness (skew)
  skew = scipy.stats.skew(psd, axis=-1)
  features.append(skew)                            #14

  # Kurtosis (kurt)
  kurt = scipy.stats.kurtosis(psd, axis=-1)
  features.append(kurt)

  features.append(np.std(psd, axis=-1))


  return np.array(features).T
```
